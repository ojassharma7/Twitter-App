```python

# import json
import psycopg2

# PostgreSQL connection details
PGHOST = "twitter.postgres.database.azure.com"
PGUSER = "neeraj"
PGPORT = 5432
PGDATABASE = "postgres"
PGPASSWORD = "Dbms@123"

# Connect to PostgreSQL
conn = psycopg2.connect(
    dbname=PGDATABASE,
    user=PGUSER,
    password=PGPASSWORD,
    host=PGHOST,
    port=PGPORT
)

# Open a cursor to perform database operations
cur = conn.cursor()

# Create users table if not exists


cur.execute("""
    CREATE TABLE IF NOT EXISTS users_final (
        id BIGINT PRIMARY KEY,
        name VARCHAR(255),
        screen_name VARCHAR(255),
        location VARCHAR(255),
        url TEXT,
        followers_count INTEGER,
        friends_count INTEGER,
        favourites_count INTEGER,
        statuses_count INTEGER,
        verified BOOLEAN,
        created_at TIMESTAMP,
        profile_image_url TEXT,
        geo_enabled BOOLEAN
    )
""")

conn.commit()
conn.close()
cur.close()
```

```python
# Define the insert query with ON CONFLICT DO NOTHING
insert_query = """
    INSERT INTO users_final (
        id, name, screen_name, location, url, followers_count, 
        friends_count, statuses_count, favourites_count, verified, 
        created_at, profile_image_url, geo_enabled
    ) VALUES (
        %(id)s, %(name)s, %(screen_name)s, %(location)s, %(url)s, 
        %(followers_count)s, %(friends_count)s, %(statuses_count)s, 
        %(favourites_count)s, %(verified)s, %(created_at)s, 
        %(profile_image_url)s, %(geo_enabled)s
    )
    ON CONFLICT DO NOTHING
"""

# PostgreSQL connection details
PGHOST = "twitter.postgres.database.azure.com"
PGUSER = "neeraj"
PGPORT = 5432
PGDATABASE = "postgres"
PGPASSWORD = "Dbms@123"

# Connect to PostgreSQL
conn = psycopg2.connect(
    dbname=PGDATABASE,
    user=PGUSER,
    password=PGPASSWORD,
    host=PGHOST,
    port=PGPORT
)

# Open a cursor to perform database operations
cur = conn.cursor()

# Insert each user data into the PostgreSQL table
for user in data:
    try:
        cur.execute(insert_query, user)
    except Exception as e:
        print(f"Error inserting user {user['id']}: {e}")

# Commit the transaction
conn.commit()

# Close communication with the PostgreSQL database
cur.close()
conn.close()
```

```python
# INDEXING
import psycopg2
PGHOST = "twitter.postgres.database.azure.com"
PGUSER = "neeraj"
PGPORT = 5432
PGDATABASE = "postgres"
PGPASSWORD = "Dbms@123"

conn = psycopg2.connect(
    dbname=PGDATABASE,
    user=PGUSER,
    password=PGPASSWORD,
    host=PGHOST,
    port=PGPORT
)
# Create a cursor object
cur = conn.cursor()

# Creating index on id column
cur.execute("CREATE INDEX idx_tweets_id ON users_final(id);")

# Creating index on created_at column
cur.execute("CREATE INDEX idx_tweets_created_at ON users_final(created_at);")

# Commit the transaction
conn.commit()

# Close the cursor and connection
cur.close()
conn.close()
```
